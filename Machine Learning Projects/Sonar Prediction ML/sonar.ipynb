## Seperate Rock and Mine

```python
X = data_frame.drop(columns=60,axis=1)
Y = data_frame[60]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,shuffle=True,stratify=Y,random_state=42)
```

Shuffle = True, Shuffles the dataset
Stratify = Y, to have equal number of rock and mine in test and train data
Random_State = 

```python
print(X.shape,X_train.shape,X_test.shape)
```

## Model Training

```python
model = LogisticRegression()
```

### Train Logistic Regression Model with Training Data

```python
model.fit(X_train,Y_train)
```

```python
X_train_pred = model.predict(X_train)

training_acc = accuracy_score(X_train_pred,Y_train)
```

```python
training_acc
```

```python
X_test_pred = model.predict(X_test)

test_acc = accuracy_score(X_test_pred,Y_test)
```
